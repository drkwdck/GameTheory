# Модель Уилсона

## Теоретические сведения

### Постановка задачи

Любая модель управления запасами, в конечном счете, должна дать ответ 
на два вопроса:
 * Какое количество продукции заказывать?
 * Когда заказывать?
 
Ответ на первый вопрос выражается через размер заказа, 
определяющего оптимальное количество ресурсов, которое необходимо 
поставлять каждый раз, когда происходит размещение заказа. В зависимости от 
рассматриваемой ситуации размер заказа может меняться во времени. Ответ на 
второй вопрос зависит от типа системы управления запасами. Если система 
предусматривает периодический контроль состояния запаса через равные 
промежутки времени (например, еженедельно или ежемесячно), момент 
поступления нового заказа обычно совпадает с началом каждого интервала 
времени. Если же в системе предусмотрен непрерывный контроль состояние 
запаса, точка заказа обычно определяется *уровнем запаса*, при котором 
необходимо размещать новый заказ.
Таким образом, решение обобщённой задачи управления запасами 
определяется следующим образом:
 * В случае периодического контроля состояния запаса следует обеспечивать поставку нового количества ресурсов в объеме размера заказа через равные интервалы времени.
 * В случае непрерывного контроля состояния запаса необходимо размещать новый заказ в размере объема запаса, когда его уровень достигает точки заказа.
 

**Проще говоря, необходимо выбирать время пополнения наших запасов наиболее оптимальным образом.**

### Модель Уилсона

Рассмотрим простейшую модель, управления запасами, предполагающую отсутствие неопределенностей.


Входные параметры модели:
 * **v** - спрос. В простейшей модели все полностью прогнозируемо, интенсивность спроса (напирмер, штук в день) известна и постоянна. То есть, в единицу времени со склада уходит **v** единиц продукции
 * **k** - стоимость доставки (например, в рублях)
 * **s** - вместисмость склада (например, в штуках)
 * **b** - стоимость склада. За стоимость склада мы принимаем цену его аренды. В простейшем случае она может постоянной или зависеть от его вместимости
 * **$\tau$** - время, необходимое для доставки новой продукции на склад (например, может измеряться в днях)
 * **F** - количество денег на нашем счету
 
Выходные параметры:
 * **h** - точка заказа
 
Если физический смысл входных параметров не вызывает вопросов, то с выходным параметром **h** не все так очевидно. Что же такое точка заказа?

### Описание моделируемого процесса

#### Точка заказа

Для того, чтобы ответить на этот вопрос, опишем процесс, который мы моделируем.
Построим график, на котором отображено количество единиц продукции на складе в зависимости от времени (количество единиц продукции **Q** не является входным параметром модели, оно должно вычисляться, опираясь на входные данные модели):


Так как спрос у нас постоянный, то мы видим, как с течением времени количество продукции на складе убывает. Позже наблюдается резкий скачок (на склад привезли новую продукцию) и снова участок убывания функции.

Очевидно, что оптимальнее всего заказывать продукции в такой момент времени, чтобы она приехала к нам на склад именно тогда, когда вся старая продукция закончится. Именно поэтому график резко подскакивает от нуля к максимуму.

Но доставка требует времени ($t_д$) и это время нам надо подбрать, создав заказ заранее. И так, когда количество продукции достигает кого-то определенного значения **h**, нам необходимо делать заказ, чтобы новый товар успел подъехать к моменту, когда старый закончится. Эта величина **h** и называется *точкой заказа*. 

#### Издержки и размер заказа

Изобразим график, на котором отображены издержки и размер заказа:


На нем отображены:
 * издержки размещения заказа - для нашей модели - это входной параметр **b**
 * издержки хранения - для нашей модели - это входной параметр **k**
 * общие издержки - это сумма издержек размещения заказа и издержек хранения
 

Наша задача найти оптимальное соотношение размера заказа и издержек.

## Примеры кода

```python
import numpy as mp
import matplotlib.pyplot as plt

# Модель Уилсона
class Wilson:
    def __init__(self, v: int, k: float, s: int, b: float, tau: int, F: float):
        self.v = v
        self.k = k
        self.s = s
        self.b = b
        self.tau = tau
        self.F = F

    # Метод для отображения на графике точки заказа
    def plot_order_point(self):
        # Количество циклов пополнения склада, которое мы будем отображать
        cycles_count = 3
        
        # Изаначально склад полностью заполнен
        # Количество продукции в момент времени t = 0 равно вместимости склада
        product_curren_count = self.s
        
        days_period = (self.s // self.v + 1) 
        # Количество дней, которое мы рассмотрим
        max_days_count = cycles_count * days_period
        
        order_point_days = []
        points = []
        
        for day in range(max_days_count):
            # Если на доставку товара требуется столько же дней, сколько на его полное израсходование,
            # делаем заказ на доставку
            if self.tau == (product_curren_count // self.v + 1):
                order_point_days.append([day, product_curren_count])
            if product_curren_count == 0:
                product_curren_count = self.s
            points.append([day, product_curren_count])
            product_curren_count -= self.v

        plt.plot([point[0] for point in points], [point[1] for point in points], 'ob', label='Q(t)')
        plt.xlabel('t, дней')
        plt.ylabel('Q, штук')
        plt.plot([point[0] for point in order_point_days], [point[1] for point in order_point_days], 'or', label='точка заказа')
        plt.legend()
```

```python
model = Wilson(1, 0.1, 10, 0.5, 3, 1500.0)
model.plot_order_point()
```
